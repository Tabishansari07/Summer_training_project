```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

iris = pd.read_csv("Iris.csv")

print("Dataset Head ")
print(iris.head())

print("\n Dataset Shape ")
print(iris.shape)

print("\n Column Names  ")
print(iris.columns)


if 'Id' in iris.columns:
    iris.drop(columns='Id', inplace=True)


print("\n Data Types and Null Values  ")
print(iris.info())


print("\n Descriptive Statistics  ")
print(iris.describe())


print("\n Unique Values per Column  ")
print(iris.nunique())


print("\n Species Counts  ")
print(iris['Species'].value_counts())


print("\n Mean Values by Species  ")
print(iris.groupby('Species').mean())

plt.figure(figsize=(10, 6))
sns.heatmap(iris.drop(columns='Species').corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Feature Correlation Heatmap")
plt.tight_layout()
plt.show()

features = iris.columns[:-1]
for feature in features:
    plt.figure(figsize=(8, 4))
    sns.boxplot(x='Species', y=feature, data=iris)
    plt.title(f"{feature.title()} by Species")
    plt.tight_layout()
    plt.show()

sns.pairplot(iris, hue='Species', corner=True)
plt.suptitle("Pairwise Feature Relationships in Iris Dataset", y=1.02)
plt.show()
```
